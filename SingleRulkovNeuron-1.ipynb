```python
import matplotlib.pyplot as plt
a=3
num_iterations=4000
b=0
u=0.001
```

```python
def SingleRulkovNeuron(X, a, b):
    x, y = X
    xn = fun(x, y, a)
    yn = y - u * (x + 1) + u * b
    return xn, yn
```

```python
def fun(x, y, a):
    if x <= 0:
        return (a/(1-x)+y)
    elif (x>0) and (x <a+y):
        return (a + y)
    elif (x>=a+y):
        return (-1)
```

```python
X01 = [0, 0]
xn1=0
X1, Y1, Z1 = [], [], []
for i in range(2000):
    xn1, yn1 = SingleRulkovNeuron(X01, a, b)
    X1, Y1, Z1 = X1 + [xn1], Y1 + [yn1], Z1+[i]
    X01 = [xn1, yn1]
```

```python
X1, Y1, Z1 = [], [], []
for i in range(num_iterations):
    xn1, yn1 = SingleRulkovNeuron(X01, a, b)
    X1, Y1, Z1 = X1 + [xn1], Y1 + [yn1], Z1+[i]
    X01 = [xn1, yn1]
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(X1, Y1, color = 'blue')
plt.xlabel('x', fontsize=25)
plt.ylabel('y', fontsize=25)
plt.tick_params(labelsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
ax.plot(Z1, X1, color = 'blue')
plt.xlabel('t', fontsize=25)
plt.ylabel('x', fontsize=25)
plt.tick_params(labelsize=20)
plt.show()
```
